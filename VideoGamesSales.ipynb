```python
plt.figure(figsize=(10, 4))
sns.lineplot(dados , x = "Ano" , y = "Na_vendas" , color = 'b' , marker='o')
plt.title('Vendas nacionais por ano')
plt.xlabel('Ano')
plt.ylabel('Vendas nacionais')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.lineplot(dados , x = "Ano" , y = 'Vendas_globais' , color = 'y' ,  marker='o')
plt.title('Vendas globais por ano')
plt.xlabel('Ano')
plt.ylabel('Vendas globais')
plt.grid(True)
plt.show()
```

```python
# Selecionar apenas as colunas numéricas
colunas_numericas = dados.select_dtypes(include=['float64', 'int64']).columns
dados_numericos = dados[colunas_numericas]

# Calcular a correlação entre as variáveis numéricas
correlation_matriz = dados_numericos.corr()

# Criar uma matriz de anotações com os valores formatados
annotations = np.array([[f"{val:.2f}" for val in row] for row in correlation_matriz.to_numpy()])

# Plotar o heatmap da matriz de correlação
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matriz, annot=annotations, cmap='coolwarm', fmt='', linewidths=0.5)
plt.title('Heatmap da Matriz de Correlação')
plt.show()
```
